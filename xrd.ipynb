# Analyse du spectre de diffraction du cristal KTiO2

Dans cette tâche, nous allons déterminer les 3 premiers pics du diffractogramme correspondant au cristal KTiO2 pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

### Imports

```python
from pymatgen import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import *
from matplotlib import pyplot as plt
```

### Téléchargement du fichier Cif

```python
with MPRester("zJnMwSh4qdHWgiI1") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-8175")
    CifWriter(structure,0.1, write_magmoms=False).write_file("KTIO2_mp-9361_symmetrized.cif")
```

### Récupération du diffractogramme

```python
xrd = XRDCalculator("CuKa")
pattern = xrd.get_pattern(structure)
```

### Affichage des données des 3 premiers pics

```python
for i in range(3):
    if i == 0:
        print("Premier pic : \n")
    if i == 1:
        print("Deuxième pic : \n")
    if i == 2:
        print("Troisième pic : \n")
        
    print("Angle 2\u03B8(en \u00b0) :",pattern.x[i])
    print("Amplitude :", pattern.y[i])
    print("Distance interplanaire :",pattern.d_hkls[i])
    print("Indices de Bravais-Miller (h,k,l) :",pattern.hkls[i][0]['hkl'])
    print("Multiplicité :",pattern.hkls[i][0]["multiplicity"])
    if i !=2:
        print("\n")
        print("-----------------------------------------------------------------------")

```

### Affichage du diffractogramme

```python
xrd.show_plot(structure)
```
